# Autograd

example 1

y=x^2

```python
device=torch.device('cuda' if torch.cuda.is_available() else 'cpu')
```

```python
#requires grad=true to tell pytorch to make derivatives
x=torch.tensor(3.0,device=device,requires_grad=True)
```

```python
y=x**2
```

```python
x
```

```python
y
```

```python
y.backward()
```

```python
x.grad
```

```python
def dy_dx(x):
  return 2*x
```

```python
dy_dx(x)
```

y=x^2 and z=sin(y)

```python
def dz_dx(x):
  return 2*x*math.cos(x**2)
dz_dx(3)
```

## Making a simple neural network now

```python
X_train_tensor = torch.from_numpy(X_train)
X_test_tensor = torch.from_numpy(X_test)
y_train_tensor = torch.from_numpy(y_train)
y_test_tensor = torch.from_numpy(y_test)
```

```python
class MeraNN():

  def __init__(self,x:torch.tensor,y:torch.tensor):
          self.x=x
          self.y=y
          self.weights=torch.rand(self.x.shape[1],1,requires_grad=True,dtype=torch.float64)
          self.bias=torch.randn(1,requires_grad=True,dtype=torch.float64)

  def forwa_pass(self,x):

    z=torch.matmul(x,self.weights)+self.bias
    y_pred=torch.sigmoid(z)
    return y_pred

  def loss(self,y_pred:torch.tensor):

      y_train_tensor=self.y
      epsilon = 1e-7
      y_pred = torch.clamp(y_pred, epsilon, 1 - epsilon)
      # Calculate loss
      loss = -(y_train_tensor * torch.log(y_pred) + (1 - y_train_tensor) * torch.log(1 - y_pred)).mean()
      return loss

  def back_prop(self,loss):

    return loss.backward()


class TrainingNN(MeraNN):

  def __init__(self,X,Y):

    super().__init__(X,Y)

  def training(self,epochs:int,lr:float):

    for epocs in range(epochs):

       y_prediction=self.forwa_pass(self.x)
       loss=self.loss(y_prediction)
       backward_prop=self.back_prop(loss)

       with torch.no_grad():
            self.weights -= lr * self.weights.grad
            self.bias -= lr * self.bias.grad

      # zero gradients
       self.weights.grad.zero_()
       self.bias.grad.zero_()

        # print loss in each epoch
       print(f'Epoch: {epocs + 1}, Loss: {loss.item()}')

```

```python
model=TrainingNN(X_train_tensor,y_train_tensor)
```

```python
model.training(epochs=35,lr=0.1)
```

```python
with torch.no_grad():
  y_pred = model.forwa_pass(X_test_tensor)
  y_pred = (y_pred > 0.9).float()
  accuracy = (y_pred == y_test_tensor).float().mean()
  print(f'Accuracy: {accuracy.item()}')
```

# Now using nn module on breast dataset

```python
X_train_tensor = torch.from_numpy(X_train).float()
X_test_tensor = torch.from_numpy(X_test)
y_train_tensor = torch.from_numpy(y_train)
y_test_tensor = torch.from_numpy(y_test)
```

```python
class SimpleNN(nn.Module):
          def __init__(self,num_features):

              super().__init__()
              self.neural=nn.Sequential(
                  nn.Linear(num_features,1),
                  nn.Sigmoid()
              )

          def forward(self,features):

            output=self.neural(features)
            return output

          def loss(self,y_pred:torch.tensor,y_true:torch.tensor):

              epsilon = 1e-7
              y_pred = torch.clamp(y_pred, epsilon, 1 - epsilon)
              # Calculate loss
              loss = -(y_true * torch.log(y_pred) + (1 - y_true) * torch.log(1 - y_pred)).mean()
              return loss


```

```python
model=SimpleNN(X_train_tensor.shape[1])
epochs=40
lr=0.1
```

```python
#training

for epocs in range(epochs):

      #forward_pass
      y_pred=model(X_train_tensor)

      #loss_calculation
      loss=model.loss(y_pred,y_train_tensor)

      #backward_propagation
      loss.backward()

      with torch.no_grad():
            model.neural[0].weight -= lr * model.neural[0].weight.grad
            model.neural[0].bias  -= lr * model.neural[0].bias.grad


      #zero_bias
      model.neural[0].weight.grad.zero_()
      model.neural[0].bias.grad.zero_()

      #print
      print(f'epoch: {epocs+1}, loss:{loss.item():.2f}')
```

```python
#evaluation

with torch.no_grad():
  y_pred=model(X_test_tensor.float())
  y_pred=(y_pred>0.8).float()
  accuracy=(y_pred==y_test_tensor).float().mean()
  print(f'accuracy: {accuracy.item()}')
```

# Using built-in loss function

```python
class SimpleNN(nn.Module):
          def __init__(self,num_features):

              super().__init__()
              self.neural=nn.Sequential(
                  nn.Linear(num_features,1),
                  nn.Sigmoid()
              )

          def forward(self,features):

            output=self.neural(features)
            return output
```

```python
#decide the epochs and lr

epochs=50

lr=0.1
```

```python
#create the model
model=SimpleNN(X_train_tensor.shape[1])

for i in range(epochs):
  #forward_prop
  y_pred=model(X_train_tensor.float())

  #loss_calculation
  loss=nn.BCELoss()
  loss=loss(y_pred,y_train_tensor.float().view(-1,1))

  #backward_prop
  loss.backward()

  #optimization
  with torch.no_grad():
    model.neural[0].weight -= lr * model.neural[0].weight.grad
    model.neural[0].bias -= lr * model.neural[0].bias.grad

  #zerogradients
  model.neural[0].weight.grad.zero_()
  model.neural[0].bias.grad.zero_()

  #print
  print(f'epoch: {i+1}, loss: {loss.item():.2f}')
```

# Using built-in Optimization algo

```python
class MeraNN(nn.Module):
  def __init__(self,num_features):
    super().__init__()
    self.neural_network=nn.Sequential(
              nn.Linear(num_features,1),
              nn.Sigmoid()
          )

  def forward(self,features):
          output=self.neural_network(features)
          return output

#define lr and epochs
epochs=37
lr=0.1

#define the model
model=MeraNN(X_train_tensor.shape[1])

#define the optimizer
optimizer=torch.optim.SGD(model.parameters(),lr=lr)

#train the neural network
for epocs in range(epochs):
  #forward_prop
  y_pred=model(X_train_tensor.float())

  #calculate_loss
  loss=nn.BCELoss()
  loss=loss(y_pred,y_train_tensor.float().view(-1,1))


  #zero_gradients
  optimizer.zero_grad()

  #backward_prop
  loss.backward()

  #params_update
  optimizer.step()


  #print
  print(f'epoch: {epocs+1}, loss: {loss.item():.2f}')


```

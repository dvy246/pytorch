## Making a simple neural network now

```python
X_train_tensor = torch.from_numpy(X_train)
X_test_tensor = torch.from_numpy(X_test)
y_train_tensor = torch.from_numpy(y_train)
y_test_tensor = torch.from_numpy(y_test)
```

```python
class MeraNN():
  def __init__(self,x:torch.tensor,y:torch.tensor):
          self.x=x
          self.y=y
          self.weights=torch.rand(self.x.shape[1],1,requires_grad=True,dtype=torch.float64)
          self.bias=torch.randn(1,requires_grad=True,dtype=torch.float64)

  def forwa_pass(self,x):
    z=torch.matmul(x,self.weights)+self.bias
    y_pred=torch.sigmoid(z)
    return y_pred

  def loss(self,y_pred:torch.tensor):
      y_train_tensor=self.y
      epsilon = 1e-7
      y_pred = torch.clamp(y_pred, epsilon, 1 - epsilon)
      # Calculate loss
      loss = -(y_train_tensor * torch.log(y_pred) + (1 - y_train_tensor) * torch.log(1 - y_pred)).mean()
      return loss

  def back_prop(self,loss):
    return loss.backward()


class TrainingNN(MeraNN):

  def __init__(self,X,Y):
    super().__init__(X,Y)

  def training(self,epochs:int,lr:float):
    for epocs in range(epochs):

       y_prediction=self.forwa_pass(self.x)
       loss=self.loss(y_prediction)
       backward_prop=self.back_prop(loss)

       with torch.no_grad():
            self.weights -= lr * self.weights.grad
            self.bias -= lr * self.bias.grad

      # zero gradients
       self.weights.grad.zero_()
       self.bias.grad.zero_()

        # print loss in each epoch
       print(f'Epoch: {epocs + 1}, Loss: {loss.item()}')

```

```python
model=TrainingNN(X_train_tensor,y_train_tensor)
```

```python
model.training(epochs=35,lr=0.1)
```

```python
with torch.no_grad():
  y_pred = model.forwa_pass(X_test_tensor)
  y_pred = (y_pred > 0.9).float()
  accuracy = (y_pred == y_test_tensor).float().mean()
  print(f'Accuracy: {accuracy.item()}')
```

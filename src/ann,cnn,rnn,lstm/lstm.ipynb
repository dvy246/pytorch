# LSTM Text Generation - Data Preprocessing

This notebook covers the fundamental steps to prepare text data for a Deep Learning model (like LSTM).

**Workflow:**
1. **Load Data**: Define the raw text corpus.
2. **Tokenization**: Break text into tokens (words).
3. **Vocabulary**: Build a dictionary mapping words to unique integers.
4. **Vectorization**: Convert text sentences to numerical sequences.
5. **Sequence Generation**: Prepare input sequences (N-grams) for training.

```python
#now we will go through each sentence tokenize it into words then make its numerical representation using the vocabulary

def indexing(text,vocab):

   indexed_inputs=[]

   for words in text:
      if words in vocab:
         indexed_inputs.append(vocab[words])
         
      else:
         indexed_inputs.append(vocab['UNK'])
         
   return indexed_inputs
```

```python
#making input numerical sentences by looping each sentence tokenizing it checking each word in the vocab and if it exists in the vocab then adding it to the list else adding the UNK
input_numerical_sentences = []

for sentence in input_sentences:
  input_numerical_sentences.append(indexing(word_tokenize(sentence.lower()), vocab))
```

```python
class CustomDataset(Dataset):
    def __init__(self,x,y):
        self.x=x
        self.y=y
    
    def __len__(self):
        return self.x.shape[0]
    
    def __getitem__(self, index):
        return self.x[index],self.y[index]
        
```

```python
train_dataset=CustomDataset(x_train,y)
test_dataset=CustomDataset(x_test,y_test)
```

```python
# MPS does not support pin_memory currently, so we disable it unless CUDA is available
use_pin_memory = torch.cuda.is_available()
dataloader=DataLoader(train_dataset,batch_size=100,shuffle=True,pin_memory=use_pin_memory)
test_datasetloader=DataLoader(test_dataset,batch_size=100,pin_memory=use_pin_memory)
```

```python
class LsTM(nn.Module):
    def __init__(self, vocab_size, embedding_dim=100, hidden_size=150, dropout_rate=0.3):
        super().__init__()
        self.embeddings = nn.Embedding(vocab_size, embedding_dim)
        self.lstm = nn.LSTM(embedding_dim, hidden_size, batch_first=True)
        self.dropout = nn.Dropout(dropout_rate)
        self.fc = nn.Linear(hidden_size, vocab_size)

    def forward(self,x):
        embeddings = self.embeddings(x)
        embeddings = self.dropout(embeddings)
        intermediate_hidden_states, (final_hidden_state, final_cell_state) = self.lstm(embeddings)
        output = self.dropout(final_hidden_state.squeeze(0))
        output = self.fc(output)
        return output
  
```

```python
for epocs in range(epochs):

    loss=0
    
    for X,Y in dataloader:
        
        X,Y=X.to(device=device,dtype=torch.long),Y.to(device=device)

        #forward propagation
        output=model(X)

        #loss
        loss_value=loss_fn(output,Y)

        #backward propagation
        optimizer.zero_grad()

        loss_value.backward()

        #optimization
        optimizer.step()

        loss+=loss_value.item()
        
    print(f'Epoch {epocs+1}/{epochs}, Loss {loss}')
        
```

```python
def evaluation(model,dataloader):

    model.eval()

    total=0
    correct=0

    with torch.inference_mode():

        for X,Y in dataloader:

            X,Y=X.to(device=device,dtype=torch.long),Y.to(device=device)

            output=model(X)

            _,indices=torch.max(output,dim=1)

            total+=Y.size(0)

            correct+=(indices==Y).sum().item()

    accuracy=correct/total*100

    return accuracy

           
```

```python

def calculate_accuracy(model, dataloader, device):
    model.eval()  # Set the model to evaluation mode
    correct = 0
    total = 0

    with torch.no_grad():  # No need to compute gradients
        for batch_x, batch_y in dataloader:
            batch_x, batch_y = batch_x.to(device,dtype=torch.long), batch_y.to(device)

            # Get model predictions
            outputs = model(batch_x)

            # Get the predicted word indices
            _, predicted = torch.max(outputs, dim=1)

            # Compare with actual labels
            correct += (predicted == batch_y).sum().item()
            total += batch_y.size(0)

    accuracy = correct / total * 100
    return accuracy

# Compute accuracy
accuracy = calculate_accuracy(model, dataloader, device)
print(f"Model Accuracy: {accuracy:.2f}%")
```
